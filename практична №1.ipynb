Має бути як мінімум 2-3 кінцівки.

#### Деталі

Користувачу мають задаватись питання один за одним. Питання, які будуть розвивати хід гри. Так, наприклад, гра має початися на першому рівні, на якому користувачу треба перемогти своїх ворогів.


```python
# стати персонажа та ворогів
answers_array = []
import time

# function to delay print for some time
def delay_print(s):
    for c in s:
        print(c, end='')
        time.sleep(0.05)
    print('')
```

```python
# загальні дані про гілку діалогу
quest_index = 0

delay_print("Вітаю на першому рівні! Ти молодший принц стародавнього королівства Даркхейтс.\nКороль (твій батько) помирає на битві від рук правителя сусіднього королівства Пайнфорт. \nТвій старший брат може стати новим правителем, але він таємно зустрічається з принцессою Пайнфорта.\nТи можеш вбити його, підмішавши яд у келих на вечері, та стати новим королем,щоб помститися за батька.")

print("Варіанти відповідей:\n1) 'Так, я хочу вбити брата.' \n2) 'Ні, я не хочу бути вбивцею.'")

answer = input("Введіть відповідь: ")

match answer:
    case "1": delay_print("Вітаю! Тепер ти повноправний король королівства Даркхейтс")
    case "2": delay_print("Тобі треба придумати новий план помсти")
    case _: delay_print("Такого варіанту не існує, вибирай нормально")

answers_array.append(answer)
```

```python
# загальні дані про гілку діалогу
quest_index = 1
previous_answer = answers_array[quest_index - 1]

match previous_answer:
    case "1": 
        delay_print("Другий рівень. Тепер тобі треба зібрати свою армію для перемоги над королем Пайнфорта")
        print("Ти звернешся до:\n1) 'Магів'\n2) 'Королівського війська'\n3) 'Титанів'")

        answer = input("Введіть відповідь: ")

        match answer:
            case "1": 
                delay_print("Маги погодилися вступитись за королівство. Ви виграли битву та вбили короля Пайнфорта")
            case "2":
                delay_print("Королівське військо пало в бою через свою слабкість. Ваше королівство програло.")
            case "3":
                delay_print("Титани погодилися і допомогли тобі помститися за твого батька. Твоє королівство перемогло")
            case _: delay_print("Такого варіанту відповіді немає.")

    case '2':
        print("Другий рівень. Так як ти залишився всього лише принцем Даркхейтсу, то варіантів для помсти небагато")
        print("Ти можеш:\n1) 'Вступити в ряди королівського війська'\n2) 'Піти навчатись темної магії'\n3) 'Закохати у себе молодшу принцесу Пайнфорта'")
        answer = input("Введіть відповідь: ")
    
        match answer:
                case "1": 
                    delay_print("Ти багато тренувався, але тобі всеодно не вдалося вбити короля Пайнфорта і ти пав в бою")
                case "2":
                    delay_print("Ти став дуже сильним магом і помстився за свого батька, вбивши короля Пайнфорта")
                case "3":
                    delay_print("Королівство Пайнфорт почало тобі довіряти, чим ти скористався і вбив короля. ТИ помстився, але тебе знайшли та казнили")
                case _: delay_print("Такого варіанту відповіді немає.")
    case _: print("Такої відповіді не існує, спробуй ще раз")   

answers_array.append(answer)   
```

```python
# загальні дані про гілку діалогу

# 1 - 1 = 2
# 1 - 2 = 3
# 1 - 3 = 4
# 2 - 1 = 3
# 2 - 2 = 4
# 2 - 3 = 5


transformed_array = [int(i) for i in answers_array]
answer_sum = sum(transformed_array)

while True:
        if(answer_sum == 3):
            delay_print("Ти програв(")
            print("\n -------Кінець-------")
            break
        elif(answer_sum == 2 or answer_sum == 4):
            delay_print("Ти переміг! Вітаю!")
            print("\n -------Кінець-------")
            break
        elif(answer_sum == 5):
            delay_print("Ти майже переміг, помста вдалась, хоч і ціною твого життя!")
            print("\n -------Кінець-------")
            break
        else:
            input("...")
```
